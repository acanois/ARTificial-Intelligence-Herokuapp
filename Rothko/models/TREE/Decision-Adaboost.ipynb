# Decision Tree Model for Rothko Art

```python
from sklearn import tree
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import graphviz 
import pydotplus
import sklearn
import pickle

# For AdaBoost
from sklearn.ensemble import AdaBoostClassifier
```

## Load art data

```python
data = pd.DataFrame.from_csv('../../data/data.csv')
```

```python
data = data[['year','shannon_entropy','mean_color_r','luminance','contrast', 'contour']]
```

```python
# Bin the data into appropriate art timelines for the artist
bins = [1935, 1940, 1947, 1950, 1968, 1971]
data['year_bin']=pd.cut(data['year'], bins)
```

```python
data.head()
```

## Create the model

```python
# Create the Decision Tree Classifier
clf = tree.DecisionTreeClassifier()
```

```python
# Split the columns to separate the features from the result/target data
features = data[['shannon_entropy','mean_color_r','luminance','contrast', 'contour']]
target = data['year_bin'].astype(str)

# Split the data into training and testing sets
features_train, features_test, target_train, target_test = train_test_split(features, target, random_state=41)
```

```python
# Train the model
clf.fit(features_train, target_train)
```

```python
# Plot the Decicion Tree
dot_data = tree.export_graphviz(
   clf, out_file=None,
   feature_names=['shannon_entropy','mean_color_r','luminance','contrast', 'contour'],  
   class_names=list(target.unique()),  
   filled=True, rounded=True,  
   special_characters=True)  

# Save the graphic representation of the tree into an image file
graph = pydotplus.graph_from_dot_data(dot_data)
graph.write_png("RothkoTree.png")

# Display the tree
graph = graphviz.Source(dot_data)  
graph
```

```python
ada_discrete = AdaBoostClassifier(
    base_estimator=clf,
    learning_rate=1,
    n_estimators=400,
    algorithm='SAMME'
)
```

```python
ada_discrete.fit(features_train, target_train)
```

```python
# Score the model on the training data
clf.score(features_train, target_train)
```

```python
ada_discrete.score(features_train, target_train)
```

```python
# Predict the bins for the test data
p = clf.predict(features_test)
for i in range(0, len(p)):
    print(p[i] == target_test.tolist()[i], p[i], target_test.tolist()[i])
```

```python
# Score the model on the test data
clf.score(features_test, target_test)
```

```python
# Check the accuracy score for the predicted values
sklearn.metrics.accuracy_score(target_test,p)
```

## ADABOOST

```python
from sklearn import tree
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import graphviz 
import pydotplus
import sklearn
import pickle

# For AdaBoost
from sklearn.ensemble import AdaBoostClassifier
```

```python
data = pd.DataFrame.from_csv('../../data/data.csv')
```

```python
data = data[['year','shannon_entropy','mean_color_r','luminance','contrast', 'contour']]
```

```python
# Bin the data into appropriate art timelines for the artist
bins = [1935, 1940, 1947, 1950, 1968, 1971]
data['year_bin']=pd.cut(data['year'], bins)
```

```python
# Create the Decision Tree Classifier
clf = tree.DecisionTreeClassifier()
```

```python
# Split the columns to separate the features from the result/target data
features = data[['shannon_entropy','mean_color_r','luminance','contrast', 'contour']]
target = data['year_bin'].astype(str)

# Split the data into training and testing sets
features_train, features_test, target_train, target_test = train_test_split(features, target, random_state=41)
```

```python
# Train the model
clf.fit(features_train, target_train)
```

```python
ada_real = AdaBoostClassifier(
    base_estimator=clf,
    learning_rate=1,
    n_estimators=4000,
    algorithm='SAMME.R'
)
```

```python
ada_real.fit(features_train, target_train)
```

```python
# Score the model on the training data
clf.score(features_train, target_train)
```

```python
ada_real.score(features_train, target_train)
```

```python
tree_predict = clf.predict(features_test)
for i in range(0, len(tree_predict)):
    print(tree_predict[i] == target_test.tolist()[i], tree_predict[i], target_test.tolist()[i])
```

```python
ada_predict = ada_real.predict(features_test)
for i in range(0, len(ada_predict)):
    print(ada_predict[i] == target_test.tolist()[i], ada_predict[i], target_test.tolist()[i])
```

```python
# Check the accuracy score for the decision tree values
sklearn.metrics.accuracy_score(target_test, tree_predict)
```

```python
# Check the accuracy score for the adaboosted values
sklearn.metrics.accuracy_score(target_test, ada_predict)
```

## Save Adaboosted Model

```python
# Save the model to a file
filename = "RothkoDecisionTree.pkl"
pickle.dump(clf, open(filename, "wb"))
```

```python
# load the model from disk
loaded_clf = pickle.load(open(filename, "rb"))
```
